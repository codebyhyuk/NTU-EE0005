Importing Data Set

```python
import pandas as pd
import numpy as np
file_path = '/content/drive/MyDrive/EE0005 Mini Project/worlds2019.csv'
df = pd.read_csv(file_path)
```

Data Cleaning(Remove Individual Player Data and Unnecessary Columns)

```python
df1 = df[df.player == 'Team']#removing ind. player data
df2 = df1[['team', 'k', 'd', 'a', 'teamdragkills', 'teambaronkills', 'herald', 'wpm', 'wcpm', 'gdat10', 'xpdat10','result']]
df2
df2.to_csv('/content/drive/MyDrive/EE0005 Mini Project/worlds2019_cleaned.csv')
```

Calculating KDA Ratio & Adding to DataFrame

```python
file_path = '/content/drive/MyDrive/EE0005 Mini Project/worlds2019_cleaned.csv'
df = pd.read_csv(file_path)
df['KDARatio'] = ((df['k']+df['a'])/df['d'])
df
```

Calculating WPM+WCPM and Adding to DataFrame

```python
df['WardScoreperMin'] = ((df['wpm']+df['wcpm']))
df
```

Extracting Cleaned and Calculated DataFrame to seperate CSV file

```python
df = df[['team', 'KDARatio','teamdragkills', 'teambaronkills', 'herald', 'WardScoreperMin', 'gdat10', 'xpdat10','result']]
df
df.to_csv('/content/drive/MyDrive/EE0005 Mini Project/worlds2019_cleaned_calculated.csv')
df
```
